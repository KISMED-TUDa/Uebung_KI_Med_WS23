## 1. Matrizenrechnung


Gegeben sei die Matrix 
$\boldsymbol X=\begin{bmatrix} 1 & -1 & 0\\ 2 & 1 & -3\end{bmatrix} \in\mathbb{R}^{2\times 3}$. 

**a)** Berechne per Hand die Eigenwerte und Eigenvektoren der Matrizen $\boldsymbol{X}\boldsymbol X^H$ und $\boldsymbol{X}^H\boldsymbol X$ und gib die Eigenwertzerlegungen an.

$\boldsymbol{X}\boldsymbol{X}^H=\begin{bmatrix} 1 & -1 & 0 \\ 2 & 1 & -3\end{bmatrix}\begin{bmatrix} 1 & 2 \\ -1 & 1 \\ 0 & -3\end{bmatrix} = \begin{bmatrix} 2 & 1 \\ 1 & 14\end{bmatrix}$

$\boldsymbol{X}^H\boldsymbol{X}=\begin{bmatrix} 1 & 2\\ -1 & 1 \\ 0 & -3\end{bmatrix}\begin{bmatrix} 1 & -1 & 2 \\ 2 & 1 & -3\end{bmatrix} = \begin{bmatrix} 5 & 1 & -6 \\ 1 & 2 & -3 \\ -6 & -3 & 9\end{bmatrix}$ 

**Eigenwertberechnung $\boldsymbol{X}\boldsymbol{X}^H$:**

$\det(\boldsymbol{X}\boldsymbol{X}^H-\lambda\boldsymbol{I})=\det\left(\begin{bmatrix} 2-\lambda & 1 \\ 1 & 14-\lambda\end{bmatrix}\right)\overset{!}{=}0$

$(2-\lambda)(14-\lambda)-1\cdot1=28-2\lambda-14\lambda+\lambda^2-1=\lambda^2-16\lambda+27\overset{!}{=}0$

$\lambda_1=1.917$, $\lambda_2 = 14.083$

**Eigenvektorberechnung:**

$\begin{bmatrix} 2-\lambda_1 & 1 \\ 1 & 14-\lambda_1\end{bmatrix}\begin{bmatrix} x_1 \\ y_1\end{bmatrix}\overset{!}{=}\begin{bmatrix} 0 \\ 0\end{bmatrix}\to\begin{bmatrix} x_1 \\ y_1\end{bmatrix}=\begin{bmatrix} -0.997 \\ 0.082 \end{bmatrix}$

$\begin{bmatrix} 2-\lambda_2 & 1 \\ 1 & 14-\lambda_2\end{bmatrix}\begin{bmatrix} x_2 \\ y_2\end{bmatrix}\overset{!}{=}\begin{bmatrix} 0 \\ 0\end{bmatrix}\to\begin{bmatrix} x_2 \\ y_2\end{bmatrix}=\begin{bmatrix} 0.082 \\ 0.997 \end{bmatrix}$

**Eigenwertberechnung $\boldsymbol{X}^H\boldsymbol{X}$:**

$\det(\boldsymbol{X}^H\boldsymbol{X}-\mu\boldsymbol{I})=\det\left(\begin{bmatrix} 5-\mu & 1 & -6 \\1 & 2-\mu & -3\\ -6 & -3 & 9-\mu\end{bmatrix}
\right)\overset{!}{=}0$

$\mu_1 = 0$, $\mu_2=1.917$, $\mu_3 = 14.083$

**Eigenvektorberechnung:**

$\begin{bmatrix} 5-\mu & 1 & -6 \\1 & 2-\mu & -3\\ -6 & -3 & 9-\mu\end{bmatrix}\begin{bmatrix} x_1 \\ y_1 \\ z_1\end{bmatrix}\overset{!}{=}\begin{bmatrix} 0 \\ 0 \\ 0\end{bmatrix}\to\begin{bmatrix} x_1 \\ y_1 \\ z_1\end{bmatrix}=\begin{bmatrix} -0.577 \\ -0.577 \\ -0.577 \end{bmatrix}$, $\begin{bmatrix} x_2 \\ y_2 \\ z_2\end{bmatrix}=\begin{bmatrix} 0.600 \\ -0.779 \\ 0.179 \end{bmatrix}$, $\begin{bmatrix} x_3 \\ y_3 \\ z_3\end{bmatrix}=\begin{bmatrix} -0.553 \\ -0.243 \\ 0.797 \end{bmatrix}$



**b)** Berechne jetzt die Kovarianz von $\boldsymbol X$: $Cov[\boldsymbol X] = \boldsymbol\Sigma = E\Big[(\boldsymbol X-E[\boldsymbol X]) (\boldsymbol X-E[\boldsymbol X])^H\Big]$ 

Da die Zeilen von $\boldsymbol X$ mittelwertfrei sind, gilt:

$Cov[\boldsymbol X] = \boldsymbol\Sigma = E[\boldsymbol X \boldsymbol X^H] = E\begin{bmatrix} 2 & 1\\ 1 & 14\end{bmatrix}=\begin{bmatrix}2 & 1\\ 1 & 14\end{bmatrix}$


**c)** Bestimme die Eigenwertzerlegung der in b) berechneten Kovarianz und gib die Ladungsmatrix $\boldsymbol\Gamma$ der Matrix $\boldsymbol X$ an. Überprüfe Dein Ergebnis, indem Du mit Python die Singulärwertzerlegung von $\boldsymbol X$ berechnen. Was stellst du fest?

Die Eigenwertzerlegung von $Cov[\boldsymbol X] = E[\boldsymbol X \boldsymbol X^H]$ haben wir bereits in a) berechnet. Wichtig ist, die Eigenwerte von $\boldsymbol X \boldsymbol X^H$ absteigend der Größe nach zu ordnen und auch die Eigenvektoren entsprechend zu sortieren. 

Die geordneten Eigenvektoren von $\boldsymbol X\boldsymbol X^H$ bilden dann die Ladungsmatrix $\boldsymbol\Gamma$ von $\boldsymbol X$:

$\boldsymbol\Sigma = \boldsymbol\Gamma\boldsymbol\Lambda\boldsymbol\Gamma^T = \begin{bmatrix} 0.082 & -0.997\\ 0.997 & 0.082\end{bmatrix}\begin{bmatrix} 14.083 & 0  \\ 0 & 1.917 \end{bmatrix}\begin{bmatrix} 0.082 & 0.997\\ -0.997 & 0.082\end{bmatrix}$


Die Multiplikation der Datenmatrix $\boldsymbol X$ mit der Ladungsmatrix $\boldsymbol\Gamma$ entspricht einer linearen Transformation in ein Koordinatensystem, in dem die Achsen möglichst viele Informationen zusammenfassen: $\boldsymbol Y = \boldsymbol\Gamma^T\boldsymbol X$. 

Man kann zeigen, dass $Cov[\boldsymbol Y]=\boldsymbol\Lambda$. Die Werte in $\boldsymbol\Lambda$ sind also ein Maß für den Anteil der entsprechenden Hauptkomponenten an der Varianz der Gesamtdaten. Daher sind die Achsen des neuen Koordinatensystems nach absteigendem Informationsgehalt sortiert. Will man nun eine Reduktion auf z.B. 2 Hauptkomponenten, so wählt man die ersten zwei Spalten der Matrix $\boldsymbol\Gamma$ für die Transformation.

```python
import numpy as np
X = np.array([[1, -1, 0], [2, 1, -3]])
np.cov(X) # numpy.cov() gewichtet standardmäßig mit (N-1), wobei N die Anzahl der Datenpunkte (Spalten) ist; man kann mit dem Attribut ddof dagegen arbeiten (vgl. empirische Varianz, VL3, Folie 28)
```

```python
import numpy as np
U,s,V_H=np.linalg.svd(X)
print('U:\n',U,'\n s:\n',s,'\n V^H:\n',V_H)
```

$\boldsymbol U\in\mathbb{R}^{2\times 2}$, $\boldsymbol V^H\in\mathbb{R}^{3\times 3}$. Die Singulärwertmatrix $\boldsymbol S$ muss daher die Form $2\times 3$ haben. Die Ausgabe besteht aus einem Vektor $\boldsymbol s$ der Länge 2. Daraus können wir $\boldsymbol S$ rekonstruieren, indem wir den Vektor diagonalisieren und anschließend leere Spalten oder Zeilen einfügen, um die gewünschte Größe zu erreichen:

```python
z = np.array([[0],[0]])
S = np.hstack([np.diag(s),z])
S
```

Es muss gelten: $\boldsymbol X = \boldsymbol U \boldsymbol S \boldsymbol V^H$. Damit können wir die Zerlegung auf ihre Richtigkeit überprüfen:

```python
X_check = np.matmul(U,np.matmul(S,V_H))
X_check
```

Die Singulärwertzerlegung von $\boldsymbol X$ hängt also mit mit den Eigenwertzerlegungen von $\boldsymbol{XX^H}$ und $\boldsymbol{X^HX}$ zusammen: 


*   Die Eigenwerte von $\boldsymbol{XX}^H$ sind die Einträge auf der Diagonalen von $\boldsymbol{SS}^H$
*   Die Eigenwerte von $\boldsymbol{X}^H\boldsymbol X$ sind die Einträge auf der Diagonalen von $\boldsymbol{S}^H\boldsymbol S$
*   Die Eigenvektoren von $\boldsymbol{XX}^H$ bilden die Matrix $\boldsymbol U$
*   Die Eigenvektoren von $\boldsymbol{X}^H\boldsymbol X$ bilden die Matrix $\boldsymbol V$

```python
# Eigenwertzerlegung von XXH, gebildet aus U und S
XXH = np.matmul(U,np.matmul(np.matmul(S,S.transpose()),U.transpose()))
print(XXH,'\n')

# Eigenwertzerlegung von XHX, gebildet aus V und S
XHX = np.matmul(V_H.transpose(),np.matmul(np.matmul(S.transpose(),S),V_H))
print(XHX)
```

##2. Principal Component Analysis (PCA)

Die Hauptkomponentenanalyse oder Principal Component Analysis (PCA) ist ein Verfahren zur Strukturierung von großen Datensätzen mit vielen Features. Diese Strukturierung kann in erster Linie für zwei Anwendungen genutzt werden:

*   Vereinfachung der Klassifikation von Daten 
*   Veranschaulichung höherdimensionaler Matrizen

Im Folgenden werden wir eine PCA auf einem Datensatz zu chronischen Nierenerkrankungen durchführen (chronic kidney diseases, *ckd*). Die 26 Spalten enthalten 24 Features, eine ID und eine Klassifikation. Jede Zeile steht für eine untersuchte Person. Die Quelle des Datensatzes und weitere Informationen zu den Features findest Du [hier](https://archive.ics.uci.edu/ml/datasets/Chronic_Kidney_Disease).

Führe die folgenden zwei Zellen aus, um den Datensatz `kidney_disease.csv` zu importieren und zu bereinigen. Die Bereinigung beinhaltet die Löschung aller Zeilen mit mindestens einem `NaN`-Wert, sowie der Spalte `id`, die keinen Informationsgehalt hat. Außerdem werden die kategorischen Features (z.B. das Feature `appet`, das den Appetit der Person als *good* oder *poor* beschreibt), in numerische Werte transformiert. Zuletzt wird das Dataframe `kidney_dataset` noch in die Featurematrix `X` und den Zielvektor `true_label` aufgeteilt.

```python
from google.colab import drive
drive.mount('/content/gdrive')

import pandas as pd
kidney_dataset = pd.read_csv('/content/gdrive/MyDrive/kidney_disease.csv')
kidney_dataset
```

```python
# Bereinigung des Datensatzes von NaN-Werten und der Spalte id
kidney_dataset = kidney_dataset.dropna()
kidney_dataset = kidney_dataset.drop(columns='id')

# Umwandlung der kategorischen Features in Boolean mithilfe von Dictionaries und der replace-Funktion von Pandas
# rbc, pc: abnormal = 1, normal = 0
trans1={'abnormal': True, 'normal': False}
# pcc, ba: present = 1, notpresent = 0
trans2={'present': True, 'notpresent': False}
# htn, dm, cad, pe, ane: yes =1, no = 0
trans3={'yes': True, 'no': False}
# appet: good=1, poor = 0
trans4={'good': True, 'poor': False}
kidney_dataset[['rbc','pc']]=kidney_dataset[['rbc','pc']].replace(trans1)
kidney_dataset[['pcc','ba']]=kidney_dataset[['pcc','ba']].replace(trans2)
kidney_dataset[['htn','dm','cad','pe','ane']]=kidney_dataset[['htn','dm','cad','pe','ane']].replace(trans3)
kidney_dataset['appet']=kidney_dataset['appet'].replace(trans4)

# Erstellen eines eigenen Dataframes für die wahren labels
true_label = kidney_dataset['classification']
# Löschen der wahren labels aus dem Featureframe -> die wahren labels sollen nicht weiter vorverarbeitet werden!
X=kidney_dataset.drop(columns='classification')

kidney_dataset
```

Zur Berechnung der PCA muss jedes Feature standardisiert werden. Das heißt, dass die Features mittelwertfrei sein und eine Standardabweichung von 1 haben müssen - man nimmt also an, dass sie standardnormalverteilt sind. In unserem Fall heißt das, dass Mittelwert und Standardabweichung für jede **Spalte** der Matrix $\boldsymbol X$ berechnet und normiert werden müssen, denn jede Spalte steht für ein Feature - anders als in der VL oder Aufgabe 1! Dies Skalierung könnt ihr beispielsweise mit der Bibliothek `sklearn` machen, die dafür über den `StandardScaler` verfügt. Genausogut könnt ihr aber auch einfach Mittelwerte berechnen und subtrahieren, sowie die Standardabweichung berechnen und durch sie teilen. Es ergibt sich die Matrix $\tilde{\boldsymbol X}$. 

**Achtung:** Mehrere der Spalten sind vom Typ `Object`, was keine manuelle Rechnung zulässt (der `StandardScaler` hingegen hat damit keine Probleme). Mit dem Befehl `pd.to_numeric` kann man sie in einen numerischen Datentyp ändern. 

**a)** Erstelle das DataFrame `X_tilde`, das die standardisierten Features enthält!

**Option 1: Sklearn**

```python
from sklearn.preprocessing import StandardScaler
# Der StandardScaler macht alle Spalten mittelwertfrei
# with_std steht standardmäßig auf True, man muss dieses Attribut also gar nicht setzen - aber man sollte sich bewusst sein, dass das ausgeführt wird.
xtilde_sklearn = pd.DataFrame(StandardScaler(with_std=True).fit_transform(X),columns=X.columns) 

xtilde_sklearn
```

**Option 2: Manuelle Skalierung**

```python
# Ausgabe der Datentypen
X.dtypes
```

```python
# Umwandlung der object-Daten in numerische
X['pcv'] = pd.to_numeric(X['pcv'])
X['wc'] = pd.to_numeric(X['wc'])
X['rc'] = pd.to_numeric(X['rc'])
# Berechnung von spaltenweisem Mittelwert und Standardabweichung, dann Skalierung
X_tilde = X-X.mean()
X_tilde = X_tilde/np.sqrt(X.var())

X_tilde
```

**b)** Führe dann die PCA mit zwei Komponenten aus. Füge das erhaltene DataFrame mit den *true labels* zusammen und erstelle einen zweidimensionalen Scatterplot der Klassifikation über den Komponenten. Zum Plotten stellen wir euch in der folgenden Zelle die Funktion `draw_pca_scatterplot(pca_dataframe, figax, pal=palette)` zur Verfügung, die ihr verwenden könnt.


```python
import matplotlib.pyplot as plt
import seaborn as sns
palette = sns.color_palette("colorblind")

def draw_pca_scatterplot(pca_dataframe, figax, pal=palette):
  """
  Zeichne einen Scatterplot des Dataframes pca_dataframe

  :param pca_dataframe: Zu plottendes Dataframe, muss 2 Features mit Namen 'principal component 1' und 'principal component 2' und das Label mit Namen 'classification' enthalten;
                        falls classifier=True, muss außerdem eine Spalte 'prediciton' vorhanden sein
  :param figax: Die Figure oder Achse, auf der zu plotten ist
  :param pal: Die Farbpalette, mit der geplottet werden soll
  """
  targets = ['ckd','notckd']
  colors = [pal[0],pal[1]]

  figax.set_xlabel('Principal Component 1', fontsize = 15)
  figax.set_ylabel('Principal Component 2', fontsize = 15)
  for target, color in zip(targets,colors):
      indicesToKeep = pca_dataframe['classification'] == target
      figax.scatter(pca_dataframe.loc[indicesToKeep, 'principal component 1'], pca_dataframe.loc[indicesToKeep, 'principal component 2'], color = color, s = 50)
  figax.legend(targets)
  figax.grid()
```

**Möglichkeit 1: PCA mit Sklearn**

```python
from sklearn.decomposition import PCA
# Erstellen eines PCA-Objektes, Spezifikation der Anz. der Hauptkomponenten
pca = PCA(n_components=2)
# Berechnung der Hauptkomponenten
# Die Funktion fit_transform() erwartet die Daten in der für DataFrame üblichen Darstellung (jede Zeile ein Datenpunkt, jede Spalte ein Feature);
# eine Transponierung ist daher nicht nötig
principal_components = pca.fit_transform(xtilde_sklearn)
# Zusammenfügen der Hauptkomponenten mit der GT
principal_dataframe = pd.DataFrame(data = principal_components, columns = ['principal component 1', 'principal component 2'],index=true_label.index)
final_df_sklearn = pd.concat([principal_dataframe,true_label],axis=1)
final_df_sklearn
```

**Option 2: Berechnen der SVD mithilfe von Numpy**

Numpy besitzt die Unterbibliothek `linalg` für Operationen der linearen Algebra. Darüber lässt sich eine SVD der Matrix $\bar{\boldsymbol X}$ ausführen. Die daraus erhaltenen Matrizen $\boldsymbol U$, $\boldsymbol S$ und $\boldsymbol V$ können für die PCA weiterverwendet werden: 

$\boldsymbol Y = \left[\boldsymbol U^H\bar{\boldsymbol X}\right]^H = \left[\boldsymbol S\boldsymbol V^H\right]^H$.

Hierbei müssen wir unsere von der Vorlesung abweichende Definition von $\boldsymbol X$ berücksichtigen und verwenden daher $\bar{\boldsymbol X}=\tilde{\boldsymbol X}^H$. Ohne diese Transponierung wären $\boldsymbol U$ und $\boldsymbol V$ vertauscht und es würde $\boldsymbol Y = \boldsymbol X\boldsymbol V$ gelten. Die zweite Transponierung, nach der Multiplikation von $\boldsymbol U$ und $\bar{\boldsymbol X}$, sorgt dann dafür, dass $\boldsymbol Y$ so aufgebaut ist wie die ursprüngliche Datenmatrix $\boldsymbol X$.

```python
# SVD mithilfe von numpy
X_bar = X_tilde.T
U,s,VH = np.linalg.svd(X_bar)
# Diagonalisieren von s und einfügen von leeren Spalten/Zeilen
z = np.zeros((U.shape[1],VH.shape[1]-U.shape[1]))
S = np.hstack([np.diag(s),z])

# Berechnung der Hauptkomponenten über U und X und Beschränkung auf 2 Komponenten
PC = np.dot(U.T,X_bar).T[:,0:2]
# Berechnung der Hauptkomponenten über S und V und Beschränkung auf 2 Komponenten
PC2 = np.dot(S,VH).T[:,0:2]

# Überprüfen, dass beide Möglichkeiten, die PC zu berechnen zum selben Ergebnis führen
# np.allclose() wird verwendet, um Probleme durch die Rundung von Gleitkommazahlen zu vermeiden
np.allclose(PC, PC2)
```

```python
# Zusammenfügen der Hauptkomponenten mit der GT
# Die Spezifikation des Index ist nötig, damit die Zeilen richtig zuordnet werden
pc_df = pd.DataFrame(data = PC, columns = ['principal component 1', 'principal component 2'],index=true_label.index)
final_df_numpy = pd.concat([pc_df,true_label],axis=1)
final_df_numpy
```

**Option 3: Berechnung der SVD von Hand über die Kovarianzmatrix**

Wie in 1 c) besprochen, entspricht die Singulärwertzerlegung der Eigenwertzerlegung der Kovarianzmatrix. Wir können diesen Schritt daher auch ohne die Verwendung von `np.linalg.svd()` durchführen:

```python
# Berechnung der Kovarianzmatrix (von der Transponierten von X!)
# np.cov hat das Attribut 'rowvar=True'. Setzt man rowvar=False, ist das effektiv dasselbe wie eine Transponierung
R = np.cov(X_bar)
R2 = np.cov(X_tilde,rowvar=False)
np.allclose(R, R2)
```

```python
# Eigenwertzerlegung der Kovarianzmatrix
cov_eigenval, cov_eigenvec = np.linalg.eigh(R)

# Sortieren der Eigenwerte absteigend nach der Größe, entsprechende Sortierung der Eigenvektoren
# Die sortierten Eigenvektoren entsprechen der Matrix V
sort_ind = np.argsort(cov_eigenval)[::-1]
sorted_eigenval = cov_eigenval[sort_ind]
sorted_eigenvec = cov_eigenvec[:,sort_ind]

# Berechnung der Hauptkomponenten über X und die Eigenvektoren
manual_pc = np.dot(X_tilde,sorted_eigenvec)[:,0:2]

# Zusammenfügen der Hauptkomponenten mit der GT
manual_df = pd.DataFrame(data = manual_pc, columns = ['principal component 1', 'principal component 2'],index=true_label.index)
final_df_manual = pd.concat([manual_df,true_label],axis=1)
final_df_manual
```

```python
# Visualisierung und Vergleich unserer drei Optionen zur PCA-Berechnung

fig = plt.figure(figsize = (26,8))

ax = fig.add_subplot(1,3,1) 
draw_pca_scatterplot(final_df_sklearn,ax)
ax.set_title('2 component PCA (sklearn)', fontsize = 20)

ax = fig.add_subplot(1,3,2)
draw_pca_scatterplot(final_df_numpy,ax)
ax.set_title('2 component PCA (numpy)', fontsize = 20)

ax = fig.add_subplot(1,3,3)
draw_pca_scatterplot(final_df_manual,ax)
ax.set_title('2 component PCA (manual)', fontsize = 20)
```
